```python
import pandas as pd
```

```python
# Inside pandas.read_csv() method skipinitialspace parameter is present using which we can skip initial 
# space present in our whole dataframe. 
# By default, it is False, make it True to remove extra space. 
df = pd.read_csv('./data/census.csv',skipinitialspace = True)
```

```python
df.columns = df.columns.str.replace("-", "_")
# df.columns = df.columns.str.replace(" ", "")
list(df.columns)
```
